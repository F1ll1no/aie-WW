с помощью функций zeros() и ones() можно создать многомерные массивы заполненные нулями и единицами соответственно

```python
zero_massive = np.zeros((3,4))
zero_massive
```

```python
zero_massive.shape
```

```python
ones_massive = np.ones((2,5))
ones_massive
```
